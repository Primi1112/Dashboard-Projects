## Data Preparation

```python
# Convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Drop 'Unnamed: 0' not needed
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.head(10)
```

```python
# Check for duplicate rows based on all columns
duplicates = df[df.duplicated()]
duplicates
```

```python
duplicates_count = df.apply(lambda x: x.duplicated().sum())
```

```python
duplicates_count
```

```python
# Calculate Revenue
df['Revenue'] = df['Quantity Ordered'] * df['Sales']
# Create a new column called "Total Cost"
df["Total Sales"] = df["Quantity Ordered"] * df["Price Each"]

# Calculate the profit for each product
df["Profit"] = df["Revenue"] - df["Total Sales"]

# Calculate the profit margin for each product
df["Profit Margin"] = df["Profit"] / df["Revenue"] * 100
```
